# **4.1 Random Variables**



*   **Definition**: Real-valued functions defined on the sample-space, are called random variables.
*   **Example**: Suppose an experiment consist of tossing of three fair coins. If random variable $Y$ represents total number of heads on the three coins, then what is the probability distribution of $Y$.

  **Solution**: Sample space is given by $\{(H,H,H),(H,H,T),(H,T,H),(H,T,T),(T,H,H),(T,T,H),(T,H,T),(T,T,T)\}$. Then function $Y$ is the mapping from this sample space to real number and mapped value is total number of heads. Hence

  \begin{align}
     P(Y=0)=P\{(T,T,T)\}=1/8 \\
     P(Y=1)=P\{(T,T,H),(T,H,T),(H,T,T)\}=3/8 \\
     P(Y=2)=P\{(T,H,H),(H,T,H),(H,H,T)\}=3/8 \\
     P\{Y=3\}=P\{(H,H,H)\}=1/8
  \end{align}

  Additionally, $Y$ can one of the values from $0$ through $3$, we must have -
  
  \begin{align}
     P(\stackrel{3}{\underset{i=0}{\cup}} \{Y=i\}) = \stackrel{3}{\underset{i=0}{\sum}} P\{Y=i\} =1
  \end{align}
  
  This is in line with what we expect from the axioms of probability.


# **4.2 Distribution Functions**



*   **Definition**: A Cumulative Distributive Function (cdf) or simply a distribution function $F$ of a random variable $X$, is defined for all real number $b$, $-\infty < b < ∞$, by -

\begin{align}
    F(b) = P\{X \leq b \}
\end{align}

*   Some properties of cdf $F$ are -

    1. $F$ is non-decreasing, i.e., if $a<b$, then $F(a) \leq F(b)$. \\
    2. $\underset{b → ∞}{lim} F(b) = 1$. \\
    3. $\underset{b → -∞}{lim} F(b) = 0$. \\
    4. $F$ is right continious. That is for any sequence $b$ and any decreasing sequence $b_n, n\geq 1$, that converges to $b, \underset{n → ∞}{lim} F(b_n) = F(b)$.

*   **Example**: The distribution function of $F$ is given by -

$F(x) = \left\{
	\begin{array}{ll}
		0  & \mbox{if } x < 0 \\
		x/2 & \mbox{if } 0 \leq x < 1 \\
    2/3 & \mbox{if } 1 \leq x < 2 \\
    11/12 & \mbox{if } 2 \leq x < 3 \\
    1 & \mbox{if } x \geq 3 \\
	\end{array}
\right. $

# **4.3 Discreate Random Variables**



*   **Definition**: A random variable that can take on at most a countable number of possible values are called discreate random variable. Probability mass function is defined as -

\begin{align}
   p(a) = P\{X=a\}
\end{align}

*   **Example**: The probability mass function of a random vriable $X$ is given by $p(i) = c λ^i /i! , i = 0,1,2,...,$ where $λ$ is some positive value. Find (a) $P\{i=0\}$ and (b) $P\{i>2\}$

  **Solution**: Since $\stackrel{\infty}{\underset{i=0}{\sum}} p(i) = 1$, we have that -

  \begin{align} c\stackrel{\infty}{\underset{i=0}{\sum}} \frac{λ^i}{i!} = 1 ⇒ c = e^{- λ}
  \end{align}

  Hence,
  
  (a). $P\{i=0\} = e^{- λ} \frac{λ^0}{0!} = e^{- λ}$ \\

  (b). $P\{i>2\} = 1 - P\{i \leq 2\} = 1 - P\{i = 0\} - P\{i = 1\} - P\{i = 2\} = 1 - e^{- λ} - λ e^{- λ} - \frac{λ^2 e^{- λ}}{2}$

*   **Cumulative Distribution Function**: CDF can be expressed in terms of $p(a)$ by -

# **4.4 Expected Value**



*   **Definition**: If $X$ is a discreate random variable having probability mass function $p(x)$, then expected value of $X$ is defined as-

\begin{align}
   E[X] = \underset{x:p(x)>0}{\sum} xp(x)
\end{align}

   In other words, expected-value of $X$ is the weighted value of $X$ where weights are the probabilities.

   Another interpretation of probability comes from the frequency based interpretation of probability, which says that expected values is the average number that is obtained after the experiment is repeated infinite number of times.

*   **Example**: Find the expected value of $X$ when we roll a fair die where $X$ is the number on the up-turned face of the die.

  **Solution**: Since $p(1) = p(2) = p(3) = p(4) = p(5) = p(6) = 1/6$. Hence, expected value of $X$ is given by -

  \begin{align}
      E[X] = 1(1/6) + 2(1/6) + 3(1/6) + 4(1/6) + 5(1/6) + 6(1/6) = 7/2
  \end{align}

# **4.6 Variance**

*   For any random varianve $X$ and its dsitribution function $F$, we want to collect as much information as possible. For example, $E(X)$ tell that a typical value of random variable $X$ will be centered around $E(X)$. But we also want to know the variance of random variables.

*   **Example**: Consider three random variables $W, Y$ and $Z$, consider their pdf as - \\

 $W = 0$ with probability 1

 $Y = \left\{
	\begin{array}{ll}
		-1 & \mbox{with probability } 1/2 \\
		1 & \mbox{with probability } 1/2\\
	\end{array}
\right. $

 $Z = \left\{
	\begin{array}{ll}
		-100 & \mbox{with probability } 1/2 \\
		100 & \mbox{with probability } 1/2\\
	\end{array}
\right. $

All of them have same expected value, but $Z$ has much larger variance than the other two.


*   **Definition**: If $X$ is a random varible with mean $μ$, then the variance of $X$, denoted by $Var(X)$, is defined by -

    \begin{align}
       Var(X) = E[(X- \mu)^2]    
    \end{align}

The alternative formula for vriance is -

\begin{align}
   Var(X) = E[X^2] - (E[X])^2
\end{align}

*   **Example**: Calculate $Var(X)$ if $X$ represents the outcome when a fair die is rolled.

  **Solution**: We know from section 4.4 that $E(X) = 7/2$. Lets compute $E(X^2)$ -

  \begin{align}
     E[X^2] = 1^2 (1/6) + 2^2 (1/6) + 3^2 (1/6) + 4^2 (1/6) + 5^2 (1/6) + 6^2 (1/6) = 91/6   \\
     Var(X) = 91/6-(7/2)^2 = 35/12 \end{align}



*   **Proposition**: $Var(aX+b) = a^2 Var(X)$


*   **Remark**: The square-root of variance is called standard-deviation and it is represented by -

\begin{align}
   SD(X) = \sqrt{Var(X)}
\end{align}









# **4.7 Bernoulli and Binomial Random variables**



*   **Definition**: A random variable $X$ is called a Bernoulli random variable if its probability mass function is given by -

\begin{align}
   p(0) = P\{X=0\} = 1-p \\
   p(1) = P\{X=1\} = p
\end{align}
*   **Definition**: A random variable $X$ is said to be a binomial random variable with parameter $(n,p)$, if there are $n$ independent trials each of which can result into a success and a failure with probability $p$ and $(1-p)$. Its pdf is given by -

\begin{align}
   p(i) = \binom{n}{i} p^i (1-p)^{n-i}
\end{align}

To show that it is a valid pdf, we show that -

\begin{align}
   \stackrel{\infty}{\underset{i=0}{\sum}} p(i) = \stackrel{n}{\underset{i=0}{\sum}} \binom{n}{i} p^i (1-p)^{n-i} = [p+(1-p)]^n = 1
\end{align}

  **Remark**: Bernoulli random variable is a special case of binomial random variable with parameters $(1,p)$


*   **Example**: Five fair coins are flipped and outcomes are independent, find the probability mass function of the number of heads obtained.

  **Solution**: If $X$ is the number of heads that appear on the five tosses then $X ∼ Bin(5,1/2)$. Hence,

  \begin{align}
      P\{X=0\} = \binom{5}{0} \binom{1}{2}^0 \binom{1}{2}^5 = 1/32 \\
      P\{X=1\} = \binom{5}{1} \binom{1}{2}^1 \binom{1}{2}^4 = 5/32 \\
      P\{X=2\} = \binom{5}{2} \binom{1}{2}^2 \binom{1}{2}^3 = 10/32 \\
      P\{X=3\} = \binom{5}{3} \binom{1}{2}^3 \binom{1}{2}^2 = 10/32 \\
      P\{X=4\} = \binom{5}{4} \binom{1}{2}^4 \binom{1}{2}^1 = 5/32 \\
      P\{X=5\} = \binom{5}{5} \binom{1}{2}^5 \binom{1}{2}^0 = 1/32 \\
  \end{align}


*   **Properties of Binomial Random Variable**: If $X$ is a Binomial Random Variable with parameters $n$ and $p$, then -

\begin{align}
    E[X] = np \\
    Var[X] = np(1-p)
\end{align}
*   **Proposition**: If $X$ is a binomial random variable with parameters $(n,p)$, then $k$ goes from $0$ to $n$, $P\{X=k\}$ first increases monotonically and then decreases monotically, reaching its largest value when $k$ is the largest integer less than or equals to $(n+1)p$.





# **4.8 Poisson Random Variable**



*   **Definition**: A random variable $X$ takes one of the following values $0,1,2,3,...$ is said to be *Poission* random variable with parameter $λ$ if for $λ > 0$, we have -
\begin{align}
  p(i) = P\{X=i\} = \frac{e^{- λ} λ^i}{i!} \text{ } i = 0,1,2,3,...
\end{align}

To check if the above function is a valid pdf, we show that -

\begin{align}
    \stackrel{\infty}{\underset{i=0}{\sum}} p(i) = e^{- λ} \stackrel{\infty}{\underset{i=0}{\sum}} \frac{λ^i}{i!} = 1
\end{align}
*   Some examples of random variables that follows Poisson distribution are -

   1. The number of misprints on a page or a group of pages
   2. The number of people living to 100 years in a community
   3. The number of wrong phone numbers dialed in a day


*   **Example**: Suppose the number of typographical errors on a single page of this book has a poisson distribution with parameter $λ = 1/2$. Calculate the probability that there is at least one error on this page.

   **Solution**: Let $X$ denote the number of errors on this page, we have

   \begin{align}
      P\{X \geq 1 \} = 1 - P\{X = 0 \} = 1 - e^{-1/2} ≈ 0.393
   \end{align}
*   The expected value and variance of a Poisson random varible are both equals to parameter $λ$.


*   Poisson with parameter $np$, where $n$ very large, is a very good approximator for number of success in $n$ independent trials when each trial has probability of success as $p$, and $p$ is very small.

   It is noticed that poisson is a good estimator even when there is a weak dependence.


*   Poisson distribution are mostly used in the case where event occur in certain point in time and following assumptions are satisfied -

   1. Probability that exactly one event occur in a given interval of length $h$ is equals to $λ h + o(h)$, where $o(h)$ stands for any function $f(h)$ that is such that $\underset{h → 0}{lim} \frac{f(h)}{h} = 0$

   2. The probability that 2 or more events occur in an interval of length $h$ is equals to $o(h)$.

   3. For any integer $n, j_1, j_2,...,J_n$ and any set of $n$ overlapping intervals, if we define $E_i$ to be exactly $j_i$ of the events under consideration occur in the $i^{th}$ interval, then events $E_1, E_2,...,E_n$ are independent.
  
*   **Example**: Suppose that earthquakes occur in the western portion of United States in accordance with assumption 1,2 and 3 with $λ=2$ and with 1 week as the unit of time.

   (a). Find the probability that at least 3 earthquakes occur during the next 2 weeks?

   **Solution**: $P\{N(2)\geq 3\} = 1-P\{N(2) = 0\} - P\{N(2) = 1\} - P\{N(2) = 2\}$
  
   $P\{N(2)\geq 3\} = 1-e^{-4} - 4e^{-4} - \frac{4^2}{2}e^{-4} = 1-13e^{-4}$

   (b). Find the probability distribution of time, starting now, until the next earthquake.

   **Solution**: Let $X$ denote the amount of time until the next earthquake. Becuase $X$ will be greater than $t$ if and only if there is no earthquake within next $t$ units of time. This is given by -

   \begin{align}
      P\{X>t\} = P\{N(t) = 0\} = e^{-λt}
   \end{align}

   So, the cdf of $X$ is given by -

   \begin{align}
      F(t) = P\{X<t\} = 1 - P\{X>t\} = 1 - e^{-λt}
   \end{align}


*   One interesting aspect about Poisson Distribution Function is -
    
    \begin{align}
       \frac{P\{X = i+1\}}{P\{X = i\}} = \frac{e^{-λ} λ^{i+1}/(i+1)!}{e^{-λ} λ^{i}/(i)!} = frac{λ}{i+1}
    \end{align}

    This gives us the following relation -

    \begin{align}
       P\{X=1\} = λ P\{X=0\} \\
       P\{X=2\} = \frac{λ}{2} P\{X=1\} \\
       . \\
       . \\
       P\{X=i+1\} = \frac{λ}{i+1} P\{X=i\}
    \end{align}








# **4.9 Geometric Random Variable**



*   If random variable $X$ represents the number of trails required to get the first success, then pdf of $X$ is given by -

\begin{align}
   P\{X=n\} = (1-p)^{n-1}p
\end{align}

  To show that it is a valid pdf, we show that

  \begin{align}
     \stackrel{\infty}{\underset{n=1}{\sum}} P\{X=n\} = p \stackrel{\infty}{\underset{n=1}{\sum}} (1-p)^{n-1} = \frac{p}{1-(1-p)} = 1
  \end{align}
*   **Example**: An urn contains $N$ white balls and $M$ black balls. The balls are randomly selected one at a time, until a black one is obtained. If we assume that each selected ball is replaced before the next ball is drawn, what is the probability that -

   (a). Exactly $n$ draws are needed

   **Solution**: $P\{X=n\} = (\frac{N}{M+N})^{n-1} \frac{M}{M+N} = \frac{MN^{n-1}}{(M+N)^n}$

   (b). At leat $k$ draws are needed
   
   **Solution**: $P\{x \geq k\} = \frac{M}{M+N} \stackrel{\infty}{\underset{n=k}{\sum}} (\frac{N}{N+M})^{n-1}$

*   The expected value of a geometric random variable is $\frac{1}{p}$ and $Var(X) = \frac{1-p}{p}$.

# **4.11 Hypergeometric Random Variables**

If random varible $X$ represent the number of white balls to be selected, then $X$ is said to follow hypergeometric distribution with pdf as -

\begin{align}
   P\{X=i\} = \frac{\binom{m}{i} \binom{N-m}{n-i}}{\binom{N}{n}}
\end{align}
*   **Example**: A purchaser of electric components buys them in lots of size 10. It is his policy to inspect 3 components randomly and accept the lot only all the three components are nondefective. If 30% of lots have 4 defective components and 70% have only 1, what percent of the lots does the purchaser rejects?

   **Solution**: Let $A$ denote the event that the purchaser accepts the lot. Now,

   \begin{align}
      P(A) = P(A| 4 \text{ }defective) \frac{3}{10} + P(A| 1 \text{ }defective) \frac{7}{10}  = \frac{54}{100}
   \end{align}

   Hence, $46 \%$ of the lot will be rejected.

# **4.12 Zeta Distribution**



*   A random variable $X$ is said to follow zeta distribution if its probability mass fucntion is given by -

\begin{align}
   P\{X=k\} = \frac{C}{k^{α+1}}
\end{align}
for some value of $α > 0$. Since the sum of the foregoing probabilities must equal 1, it follows that -
\begin{align}
   C = [\stackrel{\infty}{\underset{k=1}{\sum}} (\frac{1}{k})^{α+1}]^{-1}
\end{align}
Zeta distribution owes its name from the Rieman Zeta Function. Zeta distribution used by Italian economist Pareto to describe the family income in a given country.
