# **2.2 Sample Space and Events**



*   Set of all possible outcomes of an experiment is called *Sample Space* and it is denoted bt $S$.
    
  Example: If the experiment is to find the gender of a new born child, then $S = \{M,F\}$
  
  Example: If the outcome of an experiment is to find all possible orders of 7 hourses that participated in a race, then $S = \{\text{all } 7! \text{ permutations of } (1,2,3,4,5,6,7)\}$
*   Any susset $E$ of the *Sample Space* is called *Event*.

  Example: If $E=\{M\}$, then child is a Male.
  
  Example: $E = $ $\{$All the outcome of the race starting with $3 \}$, that means $E$ is the event that horse 3 will win the race.

*   <u>Union of Events </u>: If there are two events, $E$ and $F$, in the *Sample Space*, then $E \cup F$ is called union of event $E$ and $F$, it contains all the elements present in either $E$ or $F$.

  Example: In the gender experiment presented above, $E \cup F = \{M,F\}$

*   <u>Intersection of Events </u>: For any two events $E$ and $F$, in the *Sample Space*, $E \cap F$ is called intersection of $E$ and $F$ and it cotains only those outcomes that are present in both $E$ and $F$.

  Example: In the gender experment, if $E = \{M\}$ and $G = \{F\}$ are two events then $E \cap F = \varnothing$ or $E \cap F = \{\}$. Becuase there is nothing common between these two events, such events are called *mutually exclusive* events.

*   We define unions and intersections of more than two events like $\stackrel{\infty}{\underset{n=1}{\cup}} E_n$ and $\stackrel{\infty}{\underset{n=1}{\cap}} E_n$

*   <u>Complement of Events </u>: All the outcomes of the *Sample Space* that are not present in the *Event*, $E$, is called complement, and is represented by $E^c$.

  Example: In the gender experiment, if event is $E = \{M\}$, then $E^c = \{F\}$

*   <u>Subset </u>: If all the elements of event $E$ are present in $F$, the $E$ is called subset of $F$, and is represented by $E âŠ‚ F$. If two sets are equal, we write them as $E = F$.

  Example: In the gender experiment example, $E = \{M\}$ and $G = \{F\}$, then $E \not\subseteq G$. But $E \subset S$.

*   Operations of union, intersection and complements follow certain rules not dissimilar to algebra. They are represented as -

  <u>Commutative Law </u>: $E \cup F = F \cup E$ and $E \cap F = F \cap E$

  <u>Associative Law </u>: $(E \cup F) \cup G = E \cup (F \cup G)$ and $(E \cap F) \cap G = E \cap (F \cap G)$

  <u>Distributive Law </u>: $(E \cup F) \cap G = (E \cap G) \cup (F \cap G)$ and $(E \cap F) \cup G = (E \cup G) \cap (F \cup G)$

  Above three laws can be verified by showing that any element that is there in the left-hand-side of the equation is also there in the right-hand-side.

*   <u>*DeMorgan's Law* </u>: It a realtion between union, intersection and complements -

  \begin{align}
      (\stackrel{n}{\underset{i=1}{\cup}} E_i)^c = \stackrel{n}{\underset{n=i}{\cap}} E_i^c  \\
      (\stackrel{n}{\underset{i=1}{\cap}} E_i)^c = \stackrel{n}{\underset{n=i}{\cup}} E_i^c
  \end{align}

*   <u>Relative Frequency Based Explanation of Concept of Probability</u>:

  \begin{align}

  \end{align}

  $n(E)$ is the number of times event $E$ has occured of $n$ repeation of the experiment. Problem with this explanation is that there is no way to determine this limiting value.
*   <u>Axiomatic Way to Deal with Probability</u>:

  We assume that for each event $E$ in the *Sample space* $S$ there exist a number $P(E)$ repesenting the probability of event $E$. We assume that $P(E)$ must satisify the following axioms -

  1. **Axiom 1**: $0 \leq P(E) \leq 1$

  2. **Axiom 2**: $P(S) = 1$

  3. **Axiom 3**: For any sequence of mutually exclusive events $E_1, E_2,...$

  \begin{align}
     P(\stackrel{\infty}{\underset{i=1}{âˆª}} E_i) = \stackrel{\infty}{\underset{i=1}{âˆ‘}} P(E_i)
  \end{align}

  **Example**: What is the probability of null event?

  **Solution**: Consider a sequence of events $E_1, E_2,...,$ where $E_1 = S$ and $E_i = \emptyset$ for $i >1$. Then -

  \begin{align}

  \end{align}

  **Example**: What is the probability of any finite sequence of events?

  **Solution**: Consider a finite sequence of events $E_1, E_2,...,E_n$ and all the events after this will be called null-event. Then, from axiom 3, we have -

  \begin{align}
     P(\stackrel{n}{\underset{i=1}{âˆª}} E_i) = \stackrel{n}{\underset{i=1}{âˆ‘}} P(E_i)
  \end{align}

# **2.5 Sample Spaces Having Equally Likely Outcomes**



*   For many experiments, we assume that all the outcomes are equally likely. That means, $P(\{1\}) = P(\{2\})= ...= P(\{n\})$

  Using Axiom 2 and 3, we get $P(\{i\})= 1/N$ and from Axiom 3 $P(\{E\})=\frac{\text{Number of points in } E}{\text{Number of points in } S}$

*   **Example**: If two dices are rolled, what is the probability that sum of uptruned faces are 7?

  **Solution**: There are 6 outcomes out of 36 possible outcomes where sum of uptrun faces are 7. Hence, desired probability is $6/36=1/6$


*   **Example**: If 3 balls are "randomly" drawn from a bowl containing 6 white balls and 5 black balls. What is the probability that one of the drawn ball is white and other 2 are black?

  **Solution**: For the case where balls are distinguishable, *Sample Space* contain $11 \times 10 \times 9 = 990$ possible outcome. Desired outcomes from this set of outcomes is given by $3 \times (6 \times 5 \times 4) = 360$. Hence, deisred probability is $360/990 = 4/11$\\

  For the case when balls are indistinguishable, the desired probabiity is given by -

  \begin{align}
     \frac{\binom{6}{1} \binom{5}{2}}{\binom{11}{3}} = 4/11
  \end{align}






# **2.6 Probability As A Continious Set Function**



*   <u>Increasing Sequence of Events</u>: A sequence of events $\{E_n, n \geq 1\}$is called an increasing sequence if $E_1 \subset E_2 \subset E_3 \subset ... \subset E_n \subset E_{n+1} \subset ...$

  And a new event is denoted by -

  \begin{align}
     \underset{n â†’ âˆž}{lim} E_n = \stackrel{\infty}{\underset{i=1}{\cup}} E_i
  \end{align}

*   <u>Decreasing Sequence of Events</u>: A sequence of events $\{E_n, n \geq 1\}$is called an increasing sequence if $E_1 \supset E_2 \supset E_3 \supset ... \supset E_n \supset E_{n+1} \supset ...$

  And a new event is denoted by -

  \begin{align}
     \underset{n â†’ âˆž}{lim} E_n = \stackrel{\infty}{\underset{i=1}{\cap}} E_i
  \end{align}

*   **<u>Proposition</u>**: If $\{E_n, n\geq 1\}$ is either decreasing or increasing sequence of events, then

\begin{align}
    \underset{n â†’ âˆž}{lim} P(E_n) = P(\underset{n â†’ \infty}{lim})
\end{align}

   **Proof**: Suppose that $\{E_n, n\geq 1\}$ is an increasing sequence and define the event $F_n, n\geq 1$ by -

   \begin{align}
       F_1 = E_1 \text{ and } F_n = E_n(\stackrel{n-1}{\underset{i=1}{\cup}} E_i)^c = E_nE_{n-1}^c \text{ for } n >1 \\
       \stackrel{\infty}{\underset{i=1}{\cup}} F_i = \stackrel{\infty}{\underset{i=1}{\cup}} E_i \text{ and }  \stackrel{n}{\underset{i=1}{\cup}} F_i = \stackrel{n}{\underset{i=1}{\cup}} E_i \text{ for } n \geq 1
   \end{align}
   Thus

   \begin{align}
      P(\stackrel{\infty}{\underset{i=1}{\cup}} E_i) = P(\stackrel{\infty}{\underset{i=1}{\cup}} F_i) = \stackrel{\infty}{\underset{i=1}{\sum}} P(F_i) = \underset{n â†’ âˆž}{lim} \stackrel{n}{\underset{i=1}{\sum}} P(F_i) = \underset{n â†’ âˆž}{lim} P(\stackrel{n}{\underset{i=1}{\cup}} F_i) = \underset{n â†’ âˆž}{lim} P(\stackrel{n}{\underset{i=1}{\cup}} E_i) = \underset{n â†’ âˆž}{lim} P(E_n)
   \end{align}
